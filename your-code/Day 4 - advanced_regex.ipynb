```
# So far we saw the main methods inside the regex package


import re

# findall method -> finds all the instances of the rule required

# search method -> find the span of the first instance of the rule found

# match -> tells you the positions of the found elements

# sub -> substitutes the found instances of your rule by something else
```

```
#just writing "words" matches the words themselves

re.findall('at',"that pilates class is at 9:00")
```

```
#and match means match precisely

re.findall('at',"At 9:00 Muna takes the bus") #At, a t do not match
```

```
#you can match multiple rules at once
re.findall('hat|hit|hot',"he took his hat off in his hut before he hit that hotdog")
```

```
#you can create a block of simbols to match -> []

re.findall('h[aio]t',"he took his hat off in his hut before he hit that hotdog")
```

```
#or a block of simbols NOT to match -> [^]

re.findall('h[^aio]t',"he took his hat off in his hut before he hit that hotdog")
```

```
#you can also just accept any character rather than listing them all -> . (except newline)

re.findall('h.t',"he took his hat off in his hut before he hit that hotdog")
```

```
#or specific substets of characters -> \w (alphanumeric), \s (whitespace), \d (digit)

display(re.findall('\w',"We1rd w@rd "))

display(re.findall('\s',"We1rd w@rd "))

display(re.findall('\d',"We1rd w@rd "))
```

```
#now, you can ask for something to repeat -> +

re.findall('lo+[nt]','Elton, this is a loooooong sentence with a looooooooooooooooot of repetition')
```

```
#or to appear once or zero times -> ?

re.findall('lo+[nt]g?','Elton, this is a loooooong sentence with a looooooooooooot of repetition')
```
