```python
superstore = pd.read_excel('/Users/arthurspriet/Desktop/Sample - Superstore-3.xls')
superstore.head()
```

```python
superstore.pivot_table(index='Category')
#Very similar to group by, also keep only the numerical values
```

```python
superstore.pivot_table(index=['Category','Region']).reset_index()
```

```python
superstore.pivot_table(index=['Category','Region'],values=['Sales','Profit'])
```

```python
superstore.pivot_table(index=['Category','Region'],values=['Sales','Profit'], aggfunc=['sum','mean'])
```

```python
superstore.pivot_table(index=['Category','Region'],values=['Sales','Profit'], aggfunc={'Sales':'sum','Profit':'mean'})
```

```python
superstore.pivot_table(index=['Category','Region'],values=['Sales','Profit'], aggfunc={'Sales':['sum','min'],'Profit':['mean','max']})




```

```python
superstore.pivot_table(index=['Category','Region'], aggfunc=['first','mean'])
```

```python
import numpy as np
```

```python
superstore.pivot_table(index=['Category','Region'],columns=['Segment'], aggfunc={'Sales':lambda x : np.max(x) - np.min(x)})
```

```python
superstore = superstore.pivot_table(index=['Category','Region'],columns = ['Segment'],
                       values = ['Profit','Sales'],
                       aggfunc={'Sales':lambda x : np.max(x) - np.min(x),'Profit':['max','min']})

superstore
```

```python
#min profit for the segment home office 
Home_office = superstore.pivot_table(index=('Segment'),values = ['Profit'], aggfunc=['min'])
Home_office
```
